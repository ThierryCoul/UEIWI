```python
import arcpy
import math
from arcpy.sa import *
import os
import pandas as pd
import numpy as np
arcpy.CheckOutExtension("Spatial")
arcpy.env.overwriteOutput = True
```

```python
arcpy.env.workspace = "Setting my Working directories."
os.chdir("Setting my Working directories.")
directory_temp = "Setting my Working directories."
for raster in arcpy.ListRasters("Landsat_Classified_RF_corrected*","TIF"):
   
    print('Define the inner variables for the Urban categorisation from the raster.')
    raster_id = raster[-8:]
    #raster_id = raster[:-4]
    year = raster_id[:4]
    urban_cat = "../Temporary_updated/Urban_cat_"+raster
    raster_output ="../Temporary_updated/built_up_"+year+".tif"
    Urban_cat_shp = "../Temporary_updated/Urban_"+year+".shp"
    Urban_cat_buffer_shp = "../Temporary_updated/Buffer_Urban_"+year+".shp"
    open_space_out_raster_tif = "../Temporary_updated/Fringe_open_space_"+year+".tif"
    urban_suburban_file_tif = "../Temporary_updated/out_raster_urban_suburban_"+year+".tif"
    open_place_fringe_file = "../Temporary_updated/Fringe_open_space_"+year+"_2.tif"
    working_dir = "../info"
    Urban_cluster_tif = "../Temporary_updated/Urban_cluster_"+year+".tif"
    Urban_cluster_shp = directory_temp+"/Urban_cluster_"+year+".shp"
    Urban_cluster_shp_2 = "../Temporary_updated/Urban_cluster_"+year+"_.shp"
    Urban_cluster_shp_project = "../Temporary_updated/Urban_cluster_"+year+"_project.shp"
    #urban_cluster_buffered = "../Temporary_updated/Urban_cluster_"+year+"_buffered.shp"
    urban_cluster_buffered =  directory_temp+"/Urban_cluster_"+year+"_buffered.shp"
    
    print("Work with the raster in "+year)
    
    print("Extracting urban values")
    out_raster = arcpy.sa.ExtractByAttributes(raster, "Value = 1")
    out_raster.save(raster_output)
    
    print("Focal statistics")
    # Geoporcessing: Focal Satistics 
    raster_output_focal = arcpy.ia.FocalStatistics(out_raster, "Rectangle 3 3 CELL", "SUM", "DATA", 1)
    
    print("Classify the urban areas")
    # Geoporcessing: Reclassify rasters 
    Urban_cat_tif = arcpy.sa.Reclassify(raster_output_focal, "Value", "1 3 1;3 5 2;5 9 3", "DATA")
    Urban_cat_tif.save(urban_cat)  

    # Geoprocessing: Reclassify
    print("Reclassify Urban areas into urban and suburban areas and leaving rural areas")
    out_raster_urban_suburban = arcpy.sa.Reclassify(Urban_cat_tif, "Value", "1 NODATA;2 2;3 3", "DATA")
    out_raster_urban_suburban.save(urban_cat)
    
    # Conversion tool: RasterToPolygon
    print("Converting the urban areas raster to polygon") 
    arcpy.conversion.RasterToPolygon(out_raster_urban_suburban, Urban_cat_shp, "NO_SIMPLIFY", "Value", "SINGLE_OUTER_PART", None)
```

```python
arcpy.env.workspace = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Input"
os.chdir("C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Input")
directory_temp = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Temporary_updated"
for raster in arcpy.ListRasters("Landsat_Classified_RF_corrected*","TIF"):
   
    print('Define the inner variables for the Urban categorisation from the raster.')
    raster_id = raster[-8:]
    #raster_id = raster[:-4]
    year = raster_id[:4]
    urban_cat = "../Temporary_updated/Urban_cat_"+raster
    raster_output ="../Temporary_updated/built_up_"+year+".tif"
    Urban_cat_shp = "../Temporary_updated/Urban_"+year+".shp"
    Urban_cat_buffer_shp = "../Temporary_updated/Buffer_Urban_"+year+".shp"
    open_space_out_raster_tif = "../Temporary_updated/Fringe_open_space_"+year+".tif"
    urban_suburban_file_tif = "../Temporary_updated/out_raster_urban_suburban_"+year+".tif"
    open_place_fringe_file = "../Temporary_updated/Fringe_open_space_"+year+"_2.tif"
    working_dir = "../info"
    Urban_cluster_tif = "../Temporary_updated/Urban_cluster_"+year+".tif"
    Urban_cluster_shp = directory_temp+"/Urban_cluster_"+year+".shp"
    Urban_cluster_shp_2 = "../Temporary_updated/Temporary_updated/Urban_cluster_"+year+"_.shp"
    Urban_cluster_shp_3 = "../Temporary_updated/Urban_cluster_"+year+"__.shp"
    Urban_cluster_shp_project = "../Temporary_updated/Urban_cluster_"+year+"_project.shp"
    urban_cluster_buffered =  directory_temp+"/Urban_cluster_"+year+"_buffered.shp"

    print("Work with the raster in "+year)          
    # Geoprocessing: Buffer analysis
    print("Making a buffer around the polygons of the raster")
    arcpy.analysis.Buffer(Urban_cat_shp, Urban_cat_buffer_shp, "100 Meters", "FULL", "ROUND", "ALL", None, "PLANAR")    

    # Geoporcessing tool: ExtractByMask
    print("Extracting land cover by mask around urban areas")
    open_space_out_raster = arcpy.sa.ExtractByMask(raster, Urban_cat_buffer_shp)
    open_space_out_raster.save(open_space_out_raster_tif)

    # Geoprocessing: arcpy.sa.Reclassify
    print("Keeping open areas only out of fringe of urban areas")
    open_place_fringe = arcpy.sa.Reclassify(open_space_out_raster_tif, "Value", "0 NODATA;1 NODATA;2 1", "DATA")
    open_place_fringe.save(open_place_fringe_file)

    # Geoprocessing: arcpy.management.MosaicToNewRaster
    print("Merging urban areas and open spaces to create urban clusters")
    factorycode = 4326
    sr = arcpy.SpatialReference(factorycode)
    input_mosaic = str(open_place_fringe_file) + ";" + str(urban_cat).split("\\")[-1]
    arcpy.management.MosaicToNewRaster(input_mosaic, working_dir, Urban_cluster_tif, sr, "8_BIT_UNSIGNED", None, 1, "LAST", "FIRST")

    # Geoprocessing: arcpy.conversion.RasterToPolygon
    print("Converting urban clusters from raster to polygons")
    arcpy.conversion.RasterToPolygon(Urban_cluster_tif, Urban_cluster_shp, "NO_SIMPLIFY", "Value", "SINGLE_OUTER_PART", None)

    # Geoprocessing: arcpy.gapro.DissolveBoundaries
    print("Dissolving adjacent polygons in the urban clusters")
    arcpy.gapro.DissolveBoundaries(Urban_cluster_shp, Urban_cluster_shp_2, "SINGLE_PART", None, None, None)

    # Geoprocessing: arcpy.management.Project
    print("Projecting the urban areas polygons with a Projected areas with equal distance reference")
    factorycode = 4088
    sr_2 = arcpy.SpatialReference(factorycode)
    arcpy.management.Project(Urban_cluster_shp_2, Urban_cluster_shp_project, sr_2, None, sr, "NO_PRESERVE_SHAPE", None, "NO_VERTICAL")

    # Geoprocessing: arcpy.management.AddGeometryAttributes
    print("Adding geometry of the surface area of all polygons in the urban clusters")
    arcpy.management.AddGeometryAttributes(Urban_cluster_shp_project, "AREA", '', "SQUARE_KILOMETERS", 'PROJCS["World_Equidistant_Cylindrical_(Sphere)",GEOGCS["GCS_Sphere_GRS_1980_Authalic",DATUM["D_Sphere_GRS_1980_Authalic",SPHEROID["Sphere_GRS_1980_Authalic",6371007.0,0.0]],PRIMEM["Greenwich",0.0],UNIT["Degree",0.0174532925199433]],PROJECTION["Equidistant_Cylindrical"],PARAMETER["False_Easting",0.0],PARAMETER["False_Northing",0.0],PARAMETER["Central_Meridian",0.0],PARAMETER["Standard_Parallel_1",0.0],UNIT["Meter",1.0]]')

    # Geoprocessing: arcpy.management.AddField
    print("Creating a new field containing the are of the buffer")
    arcpy.management.AddField(Urban_cluster_shp_project, "Buf_area", "DOUBLE", None, None, None, '', "NULLABLE", "NON_REQUIRED", '')

    # Geoprocessing: arcpy.management.CalculateField
    print("Calculating the area of the buffer")
    adjustment_value = 1
    # Dictionary to store buffer distances with feature IDs
    buffer_distances = {}

    # Function to create buffer and return the area
    def buffered_area(poly, buffer_dist):
        buffered_poly = poly.buffer(buffer_dist)
        return buffered_poly.area

    # Loop through each feature in your feature class
    print("Looking for the distance to use to make the buffers for every polygons")
    # Geoprocessing: arcpy.da.SearchCursor
    with arcpy.da.SearchCursor(Urban_cluster_shp_project, ["OID@", "SHAPE@", "POLY_AREA","Buf_area"]) as cursor:
        for row in cursor:
            feature_id = row[0] 
            original_area = row[1].area
            #print(original_area)
            desired_total_area = 1.25 * original_area

            # Implementing binary search
            low = 0
            high = 1000
            tolerance = 50

            # Iteratively adjust buffer distance
            while low <= high:
                mid = (low + high) / 2
                area_with_buffer = buffered_area(row[1], mid)

                if abs(area_with_buffer - desired_total_area) < tolerance:
                    buffer_distances[feature_id] = mid
                    break
                elif area_with_buffer < desired_total_area:
                    low = mid + adjustment_value
                else:
                    high = mid - adjustment_value

            buffer_distances[feature_id] = mid

    # Update the features with the calculated buffer distances
    with arcpy.da.UpdateCursor(Urban_cluster_shp_project, ["OID@", "Buf_area"]) as cursor:
        for row in cursor:
            feature_id = row[0]
            if feature_id in buffer_distances:
                # Update the buffer field with the calculated buffer distance
                row[1] = buffer_distances[feature_id]
                cursor.updateRow(row)

    # Geopropocessing: arcpy.analysis.Buffer
    print("Buffer around every urban cluster of 1/4 of the areas")
    arcpy.analysis.Buffer(Urban_cluster_shp_project, urban_cluster_buffered, "Buf_area", "FULL", "ROUND", "NONE", None, "PLANAR")

    # Geoprocessing: arcpy.analysis.Union
    print("Dissolving adjacent polygons in the urban clusters")
    arcpy.analysis.Union([urban_cluster_buffered], Urban_cluster_shp_2)
   
    # Geoprocessing: arcpy.gapro.DissolveBoundaries
    print("Dissolving Boundaries")
    arcpy.gapro.DissolveBoundaries(Urban_cluster_shp_2, Urban_cluster_shp_3, "SINGLE_PART", None, None, None)
```

```python
arcpy.env.workspace = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Input"
arcpy.env.overwriteOutput = True

urban_cluster_2020 = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Temporary_updated/Urban_cluster_2020__.shp"
urban_cluster_2020_2 = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Temporary_updated/Urban_cluster_2020_bis.shp"
natural_cities2020 = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Output/Cities_2020.shp"
centroids_2020 = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Temporary_updated/Centroids_2020.shp"
City_hall_locations = "D:/Working_projects/06_UN_habitat/Input/City_hall_locations_13_updated.shp"
Urban_extent = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Temporary_updated/Urban_clusters_and_urban_extent2020.shp"

arcpy.management.EliminatePolygonPart
print('Eliminate Polygon parts')
arcpy.management.EliminatePolygonPart(urban_cluster_2020, urban_cluster_2020_2, "AREA", "100 SquareKilometers", 0, "CONTAINED_ONLY")

# Geoprocessing: Spatial Join
print("Merging the city hall area with the urban extent in 2020")
arcpy.analysis.SpatialJoin(urban_cluster_2020_2, City_hall_locations, Urban_extent, "JOIN_ONE_TO_ONE", "KEEP_ALL","", "INTERSECT", "#", '')

# Geoprocessing: arcpy.analysis.Select
print("Selecting the urban clusters in 2020 based on the city hall locations")
arcpy.analysis.Select(Urban_extent, natural_cities2020, "POINT_X <> 0")

# Geoprocessing: arcpy.management.AddGeometryAttributes
factorycode = 4326
sr = arcpy.SpatialReference(factorycode)
print("Estimating the centroids of the urban clusters in 2020")
arcpy.management.AddGeometryAttributes(natural_cities2020, "CENTROID", '', '', sr)

# Geoprocessing: arcpy.management.MakeXYEventLayer
print("Making a XY layer from the centroid of te urban clusters of 2020 (1)")
arcpy.management.MakeXYEventLayer(natural_cities2020, "CENTROID_X", "CENTROID_Y", "Urban_clusters_and_urban_extent_2020_Layer", sr, None)

# Geoprocessing: arcpy.management.MakeXYEventLayer
print("Making a XY layer from the centroid of te urban clusters of 2020 (2)")
arcpy.management.CopyFeatures("Urban_clusters_and_urban_extent_2020_Layer", centroids_2020, '', None, None, None)

# Geoprocessing: Spatial Join
for year in ["2015","2010","2006"]:
    Urban_cluster_shp_2 = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Temporary_updated/Urban_cluster_"+year+"__.shp"
    natural_cities_intermediary = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Temporary_updated/Urban_clusters_and_urban_extent"+year+".shp"
    natural_cities = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Temporary_updated/Urban_clusters_and_urban_extent_with_gaps"+year+".shp"
    natural_cities_2 = "C:/Users/Coulibaly Yerema/OneDrive - Kyushu University/Work/UN_habitat/06_UN_habitat/Output/Cities_"+year+".shp"
    
    print("Merging the city center with the urban extent of "+year+" (1)")
    arcpy.analysis.SpatialJoin(Urban_cluster_shp_2, natural_cities2020, natural_cities_intermediary, "JOIN_ONE_TO_ONE", "KEEP_ALL", "#", "LARGEST_OVERLAP", "#", '')

    print("Merging the city center with the urban extent of "+year+" (2)")
    arcpy.analysis.Select(natural_cities_intermediary, natural_cities, "CENTROID_X <> 0")

    arcpy.management.EliminatePolygonPart
    print('Eliminate Polygon parts for '+year)
    arcpy.management.EliminatePolygonPart(natural_cities, natural_cities_2, "AREA", "100 SquareKilometers", 0, "CONTAINED_ONLY")
```
